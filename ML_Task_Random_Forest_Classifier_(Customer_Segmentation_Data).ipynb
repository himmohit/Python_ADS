*** Importing Library and loading Dataset**

```python
# importing important libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading dataset 

df = pd.read_csv('cs.csv')
df.head()
```

```python
# Converting categorical value in binary form.

Gender_dummies = pd.get_dummies(df.Gender, prefix='Gender')
Ever_Married_dummies = pd.get_dummies(df.Ever_Married, prefix='Ever_Married')
Profession_dummies = pd.get_dummies(df.Profession, prefix='Profession')
Spending_Score_dummies = pd.get_dummies(df.Spending_Score, prefix='Spending_Score')
Graduated_dummies = pd.get_dummies(df.Graduated, prefix='Graduated')
Var_1_dummies = pd.get_dummies(df.Var_1, prefix='Var_1')
df = pd.concat([df, Gender_dummies, Ever_Married_dummies, Profession_dummies, Spending_Score_dummies, Graduated_dummies, Var_1_dummies], axis='columns')
df.drop(['ID','Graduated','Gender','Ever_Married','Profession','Spending_Score','Var_1'], axis='columns', inplace=True)
```

*** EDA Part:- Removing unsufficient column and checking null value and removing it.**

```python
# checking null value if available in dataset

df.isnull().sum()
```

```python
# As null value found in 'Experience' column. So filling it with using mean method.

df.Work_Experience.fillna(df.Work_Experience.mean(), inplace=True)
```

```python
# As null value appear in 'Family_Size' column. So filling it using mean method.

df.Family_Size.fillna(df.Family_Size.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Creating a New Dataset skip with 'Segmentation' column because it's use in output. 

df2 = df[['Age','Work_Experience','Family_Size','Gender_Female','Gender_Male','Ever_Married_No','Ever_Married_Yes','Profession_Artist','Profession_Doctor','Profession_Engineer','Profession_Entertainment','Profession_Executive','Profession_Healthcare','Profession_Homemaker','Profession_Lawyer','Profession_Marketing','Spending_Score_Average','Spending_Score_High','Spending_Score_Low','Graduated_No','Graduated_Yes','Var_1_Cat_1','Var_1_Cat_2','Var_1_Cat_3','Var_1_Cat_4','Var_1_Cat_5','Var_1_Cat_6','Var_1_Cat_7']]
df2.head()
```

```python
# Separating features (X) here and printing it. 

X = df2
X
```

```python
# Importing 'LabelEncoder' library from sklearn. So we can categorical label (Output) column in binary form.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()


label_encoded_data = df
label_encoded_data.Segmentation = le.fit_transform(label_encoded_data.Segmentation)
label_encoded_data
```

```python
# Separating y (label) column and printing it.

y = label_encoded_data['Segmentation']
y
```

*** Importing important libraries from 'sklearn' and first Training our Model then evaluating it using RandomForestClassifier Algorithm.**

```python
# Importing 'train_test_split' from sklearn so we can split dataset in train and test part.

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
# Importing LogisticRegression from sklearn and creating a variable.

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200)
```

```python
# Fitting model.

model.fit(X_train,y_train)
```

```python
# Predicting the model on test dataset.

y_pred = model.predict(X_test)
```

```python
# checking the accuracy score.

model.score(X_train,y_train)
```
