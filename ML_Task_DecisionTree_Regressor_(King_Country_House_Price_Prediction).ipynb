```python
# Importing important Libraries

import pandas as pd
import numpy as np
```

```python
# Loading dataset using pandas

df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
# Delete irrelevant columns and again printing first 5 rows.

del df['id']
del df['date']
df.head()
```

```python
# Checking Null value in dataset

df.isnull().sum()
```

```python
# Separating features (x) and label (y) and printing features.

X = df.iloc[:,1:]
y = df.iloc[:,:1]
X
```

```python
# Printing label (y)

y
```

```python
# Importing 'train_test_split' from sklearn library and spliting data into training and test part.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Imporiting 'Decision Tree Regressor' from sklearn library.

from sklearn.tree import DecisionTreeRegressor
```

```python
# Creating a class for DecisionTree

model = DecisionTreeRegressor()
```

```python
# Fitting model with training dataset.

model.fit(X_train, y_train)
```

```python
# Checking the accuracy of our Model.

model.score(X_test, y_test)
```
